# 04 Ordinal Encoder by OrdinalEncoder class import and map function both

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Size':['s','m','l','xl','s','m','s','s','l','xl','m']})
df.head(3)
```

```python
ord_data = [['s','m','l','xl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df[['Size']])
```

```python
df['Size_en'] = oe.transform(df[['Size']])
df
```

# By Map Function

```python
ord_data1 = {'s':2,'m':3,'l':4,'xl':5}
df['Size_en_map'] = df['Size'].map(ord_data1)
df
```
