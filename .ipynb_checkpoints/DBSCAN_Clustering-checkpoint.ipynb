# 37 DBSCAN Clustering

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
import pandas as pd
```

```python
#creating non linear dataset
x,y = make_moons(n_samples=250, noise=0.05)
```

```python
df = {'data1':x[:,0], 'data2':x[:,1]} # 'Output':y
```

```python
dataset = pd.DataFrame(df)
dataset.head(3)
```

```python
sns.scatterplot(x='data1', y='data2', data=dataset)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.2, min_samples=5)
dataset['Predict'] = db.fit_predict(dataset)
```

```python
db.fit_predict(dataset) #this shows there are two clusters.
```

```python
sns.scatterplot(x='data1', y='data2', data=dataset, hue='Predict')
plt.show()
```

```python
#shows the clustering is done correct.
sns.scatterplot(x='data1', y='data2', data=dataset, hue=y)
plt.show()
```
