# 42 Bagging (Classification)

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import seaborn as sns
```

```python
x,y = make_moons(n_samples=1000,noise=0.2)
```

```python
df = {'x1':x[:,0],'x2':x[:,1],'y':y}
```

```python
dataset = pd.DataFrame(df)
```

```python
sns.scatterplot(x='x1', y='x2', data=dataset, hue='y')
plt.show()
```

```python
x_a = dataset.iloc[:,:-1]
y_a = dataset['y']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
sv = SVC()
sv.fit(x_train, y_train)
sv.score(x_train,y_train)*100
```

```python
bg = BaggingClassifier(estimator=SVC(),n_estimators=30)
bg.fit(x_train,y_train)
```

```python
bg.score(x_train, y_train)*100, bg.score(x_test, y_test)*100
```

```python
rf= RandomForestClassifier(n_estimators=30) #the overfitting case can be handled by manipulating hyper parameters
rf.fit(x_train, y_train)
rf.score(x_train,y_train)*100, rf.score(x_test,y_test)*100 
```
