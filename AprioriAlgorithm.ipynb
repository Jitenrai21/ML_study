```python
market = []
for i in range(0,dataset.shape[0]):
    cus = []
    for j in dataset.columns :
        if type(dataset[j][i]) == str:
            cus.append(dataset[j][i])
    market.append(cus)
# market -> this is in form of list inside a list
```

```python
l = []
for i in market:
    for j in i:
        # print(j)
        l.append(j)
```

```python
import collections
```

```python
p = collections.Counter(l)
```

```python
# p.keys
# p.values
```

```python
d = {'Item Name':p.keys(), 'Values':p.values()}
```

```python
pd.set_option("display.max_rows", 200)
```

```python
pd.DataFrame(d).sort_values(by=['Values'], ascending=False)
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr = TransactionEncoder()
tr.fit(market)
```

```python
df = pd.DataFrame(tr.transform(market), columns=tr.columns_)
df.head(3)
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(df, min_support=0.07, use_colnames=True, max_len=3)
```

```python
type(apriori(df, min_support=0.07, use_colnames=True, max_len=3))
```

```python
apriori(df, min_support=0.07, use_colnames=True, max_len=3).sort_values(by=['support'])
```

```python
apriori(df, min_support=0.05, use_colnames=True, max_len=3).sort_values(by=['support'])
```
