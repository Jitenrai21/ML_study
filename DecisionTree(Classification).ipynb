```python
# Decision tree works on CART algorithm CART -> Classification and Regression Tree algorithm
# Decision tree is a non linear separable algorithm.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
dataset.head(3)
```

```python
sns.scatterplot(x='Age', y= "EstimatedSalary",data=dataset, hue='Purchased')
plt.show()
```

```python
dataset.isnull().sum()
```

```python
x = dataset.drop(columns = ['Purchased'])
y = dataset['Purchased']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# to scale the values of x for closer values
sc = StandardScaler()
sc.fit(x)
x = pd.DataFrame(sc.transform(x),columns= x.columns)
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(max_depth=2) # max_depth removes the overfitting it is the pre pruning method
dt.fit(x_train, y_train)
```

```python
dt1 = DecisionTreeClassifier(criterion='entropy')
dt1.fit(x_train, y_train)
```

```python
dt.score(x_test, y_test)*100
```

```python
dt.score(x_train, y_train)*100 
```

```python
dt1.score(x_test, y_test)*100
```

```python
dt.predict([[26,43000]]) #wrong prediction
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy() ,clf=dt)
plt.show()
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy() ,clf=dt1)
plt.show()
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(50,50))
plot_tree(dt)
plt.savefig('classiDT.jpg')
plt.show()
```

# Post pruning

```python
# max depth increases in loop
# the high difference in accuracy of training set score and testing set score denotes overfitting
```

```python
for i in range(1,20):
    dt2 = DecisionTreeClassifier(max_depth=i)
    dt2.fit(x_train,y_train)
    print(dt2.score(x_train,y_train),dt2.score(x_test,y_test),i)
    #this gives the best option for max_depth depending on the lowest difference
```
