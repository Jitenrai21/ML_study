```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv(r'C:\Users\ACER\OneDrive\Desktop\AI_ML\music.csv')
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
X = df[['age', 'gender']]
Y = df['genre']
```

```python
le = LabelEncoder()
Y_encoded = le.fit_transform(Y)
```

```python
clf = DecisionTreeClassifier(criterion='entropy', max_depth = 3, random_state = 0)
clf.fit(X,Y_encoded)
```

```python
plt.figure(figsize = (12,8))
plot_tree(clf, filled=True, feature_names=['Age', 'Gender'], class_names = le.classes_)
plt.title("Decision Tree for Music genre prediction")
plt.show()
```

```python
new_sample = pd.DataFrame({'age':[28], 'gender':[1]})
prediction = clf.predict(new_sample)
predicted_genre = le.inverse_transform(prediction)
print(f"The predicted genre for Age 28 and Gender 1 is: {predicted_genre[0]}")
```
