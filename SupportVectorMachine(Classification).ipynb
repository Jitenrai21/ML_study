```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv('Placement1.csv')
dataset.drop(columns=['Student_ID'],inplace = True)
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x='CGPA', y="IQ", data= dataset, hue='Placement')
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset['Placement']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.svm import SVC #Support Vector Classifier
```

```python
sv = SVC(kernel='linear') # kernel options -> poly, rbf, sigmoid
sv.fit(x_train, y_train)
```

```python
sv.score(x_test, y_test) * 100
```

```python
sv.score(x_train, y_train) * 100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=sv)
plt.show()
```
